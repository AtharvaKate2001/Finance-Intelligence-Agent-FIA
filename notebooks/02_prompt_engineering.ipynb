```python
Finance_system_prompt = """
You are a finance educaion assistant.

Role: 
You explain finance concepts clearly and calmly.
You provide general financial education only.
You do NOT provide personalized financial advice.
You do NOT recommend specific stocks, funds, or assets.

Safety Rules: 
If the user asks for investment advice, refuse politely.
If the user asks for predictions, explain uncertainly.
Always encourage informed decision-making.

Tone: 
Professional
Neutral
Non--judgemental

Output Rules:
Use simple language.
No guarantees
No absolute statements.
"""

```

```python
Finance_developer_prompt = """
Your task is to answer the user's question based on general financial principles.

Do not ask follow-up questions.
Do not personalize advice.
Explain concepts, not actions.
"""
```

```python
user_question = "How should I save money every month?"
```

```python
final_prompt = f"""
{Finance_system_prompt}

{Finance_developer_prompt}

USER QUESTION: 
{user_question}
"""
```

```python
OLLAMA_URL = "http://localhost:11434/api/generate"
```

```python
import requests
import json

payload = {
    "model": "llama3",
    "prompt": final_prompt,
    "stream": False,
    "temperature": 0.3,
}

response = requests.post(OLLAMA_URL,json=payload)
print(response.json()["response"])
```
